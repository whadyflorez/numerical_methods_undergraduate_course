# Problema de difusión no lineal con coeficiente variable
$k(T)\frac{d^2T}{dx^2}+q=0$

```python
import numpy as np
from scipy.optimize import fsolve
import matplotlib.pyplot as plt

L=2.0
q=100.0 
N=25
h=L/(N-1)

x=np.linspace(0,L,N)

def k(T):
    y=1+0.1*T+0.01*T**2
    return y 

def eqs(T):    
    y=np.zeros(N)
    y[0]=T[0]
    for i in range(1,N-1):
        y[i]=k(T[i])*(T[i+1]-2*T[i]+T[i-1])/h**2+q
    y[N-1]=(T[N-1]-T[N-2])/h 
    return y

T_ini=np.ones(N)

T=fsolve(eqs,T_ini)

plt.plot(x,T,'o--')

    

    
```
