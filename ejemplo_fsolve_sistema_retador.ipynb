# Sistema no lineal más retador con `fsolve`

## Planteamiento del problema

Considere el siguiente sistema de tres ecuaciones no lineales con tres incógnitas:

\[
f_1(x,y,z)=e^x+y-z^2-1=0
\]

\[
f_2(x,y,z)=x^2+y^2+z^2-6=0
\]

\[
f_3(x,y,z)=x+y+z-1=0
\]

Se desea encontrar una solución numérica del sistema usando la función `fsolve` de `scipy.optimize`.

## Comentario didáctico

Este sistema es más interesante que el anterior porque mezcla:

- una función exponencial,
- términos cuadráticos,
- y una ecuación lineal.

Por tanto, el comportamiento numérico puede ser más sensible a la suposición inicial.

## 1. Importación de librerías

```python
import numpy as np
from scipy.optimize import fsolve
```

## 2. Definición del sistema

```python
def sistema(vars):
    x, y, z = vars

    f1 = np.exp(x) + y - z**2 - 1
    f2 = x**2 + y**2 + z**2 - 6
    f3 = x + y + z - 1

    return [f1, f2, f3]
```

## 3. Suposición inicial

Se propone comenzar con:

\[
(x_0,y_0,z_0)=(1,1,-1)
\]

Esta elección no garantiza la solución exacta, pero sirve como punto de partida para el algoritmo iterativo.

```python
x0 = [1.0, 1.0, -1.0]
x0
```

## 4. Solución numérica con `fsolve`

```python
solucion = fsolve(sistema, x0)
solucion
```

## 5. Presentación de resultados

```python
x, y, z = solucion

print("Solución encontrada:")
print(f"x = {x:.8f}")
print(f"y = {y:.8f}")
print(f"z = {z:.8f}")
```

## 6. Verificación de residuos

```python
residuos = sistema(solucion)

print("Residuos del sistema:")
for i, r in enumerate(residuos, start=1):
    print(f"f{i} = {r:.12e}")
```

## 7. Discusión

### Aspectos importantes

1. El sistema puede tener más de una solución.
2. La convergencia puede depender de la condición inicial.
3. En problemas no lineales más complejos, conviene ensayar varios valores iniciales.
4. Aunque aquí no damos el Jacobiano, `fsolve` puede aproximarlo internamente.
