# Sistema no lineal con `fsolve` y Jacobiano explícito

## Planteamiento del problema

Considere el sistema:

\[
f_1(x,y,z)=e^x+y-z^2-1=0
\]

\[
f_2(x,y,z)=x^2+y^2+z^2-6=0
\]

\[
f_3(x,y,z)=x+y+z-1=0
\]

Ahora se resolverá el sistema usando `fsolve`, pero además se suministrará de manera explícita la matriz Jacobiana.

## ¿Qué es el Jacobiano?

Para un sistema de \(n\) ecuaciones no lineales con \(n\) incógnitas, el Jacobiano es la matriz de derivadas parciales:

\[
J(x,y,z)=
\begin{bmatrix}
\dfrac{\partial f_1}{\partial x} & \dfrac{\partial f_1}{\partial y} & \dfrac{\partial f_1}{\partial z}\\
\dfrac{\partial f_2}{\partial x} & \dfrac{\partial f_2}{\partial y} & \dfrac{\partial f_2}{\partial z}\\
\dfrac{\partial f_3}{\partial x} & \dfrac{\partial f_3}{\partial y} & \dfrac{\partial f_3}{\partial z}
\end{bmatrix}
\]

En este caso:

\[
\frac{\partial f_1}{\partial x}=e^x, \qquad
\frac{\partial f_1}{\partial y}=1, \qquad
\frac{\partial f_1}{\partial z}=-2z
\]

\[
\frac{\partial f_2}{\partial x}=2x, \qquad
\frac{\partial f_2}{\partial y}=2y, \qquad
\frac{\partial f_2}{\partial z}=2z
\]

\[
\frac{\partial f_3}{\partial x}=1, \qquad
\frac{\partial f_3}{\partial y}=1, \qquad
\frac{\partial f_3}{\partial z}=1
\]

Por tanto,

\[
J(x,y,z)=
\begin{bmatrix}
e^x & 1 & -2z\\
2x & 2y & 2z\\
1 & 1 & 1
\end{bmatrix}
\]

## 1. Importación de librerías

```python
import numpy as np
from scipy.optimize import fsolve
```

## 2. Definición del sistema

```python
def sistema(vars):
    x, y, z = vars

    f1 = np.exp(x) + y - z**2 - 1
    f2 = x**2 + y**2 + z**2 - 6
    f3 = x + y + z - 1

    return [f1, f2, f3]
```

## 3. Definición del Jacobiano explícito

```python
def jacobiano(vars):
    x, y, z = vars

    J = np.array([
        [np.exp(x), 1.0, -2.0*z],
        [2.0*x, 2.0*y, 2.0*z],
        [1.0, 1.0, 1.0]
    ])

    return J
```

## 4. Punto inicial

Se tomará como punto de partida:

\[
(x_0,y_0,z_0)=(1,1,-1)
\]

```python
x0 = [1.0, 1.0, -1.0]
x0
```

## 5. Solución con `fsolve` suministrando el Jacobiano

En `fsolve`, el Jacobiano se pasa mediante el argumento `fprime`.

```python
solucion = fsolve(sistema, x0, fprime=jacobiano)
solucion
```

## 6. Resultados

```python
x, y, z = solucion

print("Solución encontrada:")
print(f"x = {x:.8f}")
print(f"y = {y:.8f}")
print(f"z = {z:.8f}")
```

## 7. Verificación de residuos

```python
residuos = sistema(solucion)

print("Residuos del sistema:")
for i, r in enumerate(residuos, start=1):
    print(f"f{i} = {r:.12e}")
```

## 8. Evaluación del Jacobiano en la solución

Esto permite discutir si la matriz está bien condicionada o no cerca de la raíz.

```python
J_sol = jacobiano(solucion)
print("Jacobiano evaluado en la solución:")
print(J_sol)
```

## 9. Discusión didáctica

### Ventajas de usar Jacobiano explícito

1. Se aprovecha información analítica del problema.
2. Puede mejorar la estabilidad y rapidez de convergencia.
3. Ayuda a que el estudiante entienda la base matemática del método de Newton para sistemas.

### Idea clave

Cuando no se suministra el Jacobiano, `fsolve` lo aproxima numéricamente.  
Cuando sí se entrega, el método puede volverse más transparente y más cercano al desarrollo teórico visto en clase.

```python
# Comparación opcional: resolver sin Jacobiano explícito
solucion_sin_J = fsolve(sistema, x0)

print("Solución sin Jacobiano explícito:")
print(solucion_sin_J)

print("\nDiferencia entre ambas soluciones:")
print(solucion - solucion_sin_J)
```
