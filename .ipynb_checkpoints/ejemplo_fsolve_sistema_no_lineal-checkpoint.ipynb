# Solución de un sistema de ecuaciones no lineales con `fsolve`

## Planteamiento del problema

Considere el siguiente sistema de **tres ecuaciones no lineales** con tres incógnitas:

\[
f_1(x,y,z)=x^2+y^2+z^2-9=0
\]

\[
f_2(x,y,z)=x+y-z-1=0
\]

\[
f_3(x,y,z)=x-y+z-2=0
\]

Se desea encontrar una solución \((x,y,z)\) que satisfaga simultáneamente las tres ecuaciones.

## Interpretación geométrica

Este sistema puede interpretarse como la intersección entre:

- una **esfera** de radio 3 centrada en el origen,
- y **dos planos** en el espacio tridimensional.

## Objetivo numérico

Resolver el sistema utilizando la función `fsolve` de `scipy.optimize`, que permite aproximar la solución numérica a partir de una suposición inicial.

## 1. Importación de librerías

En este ejemplo usaremos:

- `numpy` para manejo numérico,
- `fsolve` de `scipy.optimize` para resolver el sistema.

```python
import numpy as np
from scipy.optimize import fsolve
```

## 2. Definición del sistema de ecuaciones

La función debe recibir un vector con las incógnitas y retornar una lista con los residuos de las ecuaciones.

```python
def sistema(vars):
    x, y, z = vars

    f1 = x**2 + y**2 + z**2 - 9
    f2 = x + y - z - 1
    f3 = x - y + z - 2

    return [f1, f2, f3]
```

## 3. Selección de una suposición inicial

Los métodos tipo Newton requieren un punto inicial.  
En `fsolve`, esta suposición inicial se entrega como un vector.

En este caso usaremos:

\[
(x_0,y_0,z_0)=(1,1,1)
\]

```python
x0 = [1.0, 1.0, 1.0]
x0
```

## 4. Solución numérica con `fsolve`

Ahora resolvemos el sistema.

```python
solucion = fsolve(sistema, x0)
solucion
```

## 5. Presentación ordenada de resultados

```python
x, y, z = solucion

print("Solución encontrada:")
print(f"x = {x:.6f}")
print(f"y = {y:.6f}")
print(f"z = {z:.6f}")
```

## 6. Verificación de la solución

Para comprobar la calidad de la solución, sustituimos el resultado en las ecuaciones originales.  
Si la solución es correcta, los residuos deben ser cercanos a cero.

```python
residuos = sistema(solucion)

print("Residuos del sistema evaluado en la solución:")
for i, r in enumerate(residuos, start=1):
    print(f"f{i} = {r:.12e}")
```

## 7. Discusión didáctica

### Observaciones importantes

1. `fsolve` busca una raíz del sistema, es decir, un punto donde todas las ecuaciones valgan cero simultáneamente.

2. El resultado puede depender de la **suposición inicial**.  
   Si se cambia el valor inicial, el método podría converger a otra solución o incluso no converger.

3. Siempre es necesario verificar los **residuos**.

4. Aunque `fsolve` evita implementar manualmente el método de Newton, internamente usa ideas basadas en linealización y búsqueda iterativa de raíces.

## Conclusión

Este ejemplo muestra cómo resolver un sistema no lineal de tres ecuaciones con tres incógnitas usando herramientas numéricas de Python de manera simple y efectiva.

## 8. Ejercicio sugerido para el estudiante

Cambie la suposición inicial por otros valores, por ejemplo:

- `(2, 2, 1)`
- `(-1, 1, 2)`
- `(0, 0, 3)`

y analice:

- si el método converge,
- a qué solución converge,
- y cómo cambian los residuos.

```python
# Prueba aquí con otros valores iniciales
x0_nuevo = [2.0, 2.0, 1.0]
solucion_nueva = fsolve(sistema, x0_nuevo)

print("Nueva solución encontrada:")
print(solucion_nueva)

print("\nResiduos:")
print(sistema(solucion_nueva))
```
