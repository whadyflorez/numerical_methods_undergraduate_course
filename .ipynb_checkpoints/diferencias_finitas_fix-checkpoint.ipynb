# Cálculo de derivadas mediante diferencias finitas

## Planteamiento

Trabajaremos con la función:

$$
f(x)=e^x \sin x
$$

Derivadas exactas:

$$
f'(x)=e^x(\sin x+\cos x)
$$

$$
f''(x)=2e^x\cos x
$$

$$
f'''(x)=2e^x(\cos x-\sin x)
$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Primera derivada

### Diferencia hacia adelante

$$
f'(x) \\approx \\frac{f(x+h)-f(x)}{h} \\quad O(h)
$$

### Diferencia centrada

$$
f'(x) \\approx \\frac{f(x+h)-f(x-h)}{2h} \\quad O(h^2)
$$

### Diferencia centrada de orden 4

$$
f'(x) \\approx \\frac{-f(x+2h)+8f(x+h)-8f(x-h)+f(x-2h)}{12h}
$$

```python
def f(x):
    return np.exp(x)*np.sin(x)

def df_exact(x):
    return np.exp(x)*(np.sin(x)+np.cos(x))

def d1_center_o2(f,x,h):
    return (f(x+h)-f(x-h))/(2*h)

def d1_center_o4(f,x,h):
    return (-f(x+2*h)+8*f(x+h)-8*f(x-h)+f(x-2*h))/(12*h)
```

## Ejemplo numérico

```python
x0=1.0
hs=[0.2,0.1,0.05,0.025]

exact=df_exact(x0)

for h in hs:
    a2=d1_center_o2(f,x0,h)
    a4=d1_center_o4(f,x0,h)
    print("h=",h)
    print("O(h^2):",a2,"error:",abs(a2-exact))
    print("O(h^4):",a4,"error:",abs(a4-exact))
    print()
```

## Segunda derivada

$$
f''(x) \\approx \\frac{f(x+h)-2f(x)+f(x-h)}{h^2}
$$

```python
def d2_center(f,x,h):
    return (f(x+h)-2*f(x)+f(x-h))/(h**2)
```

## Tercera derivada

$$
f^{(3)}(x) \\approx \\frac{f(x+2h)-2f(x+h)+2f(x-h)-f(x-2h)}{2h^3}
$$

```python
def d3_center(f,x,h):
    return (f(x+2*h)-2*f(x+h)+2*f(x-h)-f(x-2*h))/(2*h**3)
```
